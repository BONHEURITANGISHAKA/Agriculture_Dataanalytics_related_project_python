## Identify Anomalies in Water Parameters

### Subtask:
Implement a method to detect unusual readings or outliers in water temperature, pH, and TDS. This will involve statistical or rule-based approaches to pinpoint critical conditions for fish health or system malfunctions, showcasing your ability to derive critical insights from IoT data.


**Reasoning**:
I will define a function `detect_outliers_iqr` that takes a pandas Series, calculates the IQR, and returns a DataFrame indicating outliers based on the lower and upper bounds. This function will be reusable for all specified columns.



```python
def detect_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    is_outlier = (series < lower_bound) | (series > upper_bound)
    return pd.DataFrame({'lower_bound': lower_bound, 'upper_bound': upper_bound, 'is_outlier': is_outlier})

print("detect_outliers_iqr function defined.")
```

**Reasoning**:
Now that the `detect_outliers_iqr` function is defined, I will apply it to the 'water_temp', 'water_ph', and 'tds' columns, store the results, and then print the number of anomalies found for each parameter, fulfilling the remaining instructions of the subtask.



```python
outliers_water_temp = detect_outliers_iqr(df['water_temp'])
outliers_water_ph = detect_outliers_iqr(df['water_ph'])
outliers_tds = detect_outliers_iqr(df['tds'])

print("Number of anomalies in Water Temperature:", outliers_water_temp['is_outlier'].sum())
print("Number of anomalies in Water pH:", outliers_water_ph['is_outlier'].sum())
print("Number of anomalies in TDS:", outliers_tds['is_outlier'].sum())
```
